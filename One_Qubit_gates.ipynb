```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector, array_to_latex, plot_histogram
from qiskit_aer import AerSimulator
from math import pi
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
def plot_bloch(circuit_name):
  simulator = Aer.get_backend('statevector_simulator')
  result = execute(circuit_name, backend=simulator).result()
  statevector = result.get_statevector()
  return plot_bloch_multivector(result.get_statevector())

def state_vector(circuit_name):
  simulator = Aer.get_backend('statevector_simulator')
  result = execute(circuit_name, backend=simulator).result()
  statevector = result.get_statevector()
  return array_to_latex(statevector, prefix="\\text{statevector = }\n")
```

```python
circuit = QuantumCircuit(3)
circuit.measure_all()
circuit.draw(output='mpl')
```

```python
state_vector(circuit)
```

```python
plot_bloch(circuit)
```

```python
aersim = AerSimulator()
# Perform an ideal simulation
result_ideal = aersim.run(circuit).result()
counts_ideal = result_ideal.get_counts(0)
plot_histogram(counts_ideal)
```

The **Pauli gates** are a set of quantum logic gates named after the physicist Wolfgang Pauli.
There are three Pauli gates: the Pauli-X gate, the Pauli-Y gate, and the Pauli-Z gate

*   The Pauli-X gate (or NOT gate) performs a *bit-flip* operation on a qubit, similar to the classical NOT gate. When applied to a qubit in the state |0⟩, it transforms it to |1⟩, and vice versa.In matrix form, it is represented as: $X = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$

*   The Pauli-Y gate is a *combination of a bit-flip and a phase-flip* operation. It transforms the state |0⟩ to i|1⟩ and the state |1⟩ to -i|0⟩. In matrix form, it is represented as: $Y = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix}$

*   The Pauli-Z gate introduces a *phase flip* without affecting the state of the qubit. It leaves the state |0⟩ unchanged and transforms the state |1⟩ to -|1⟩. In matrix form, it is represented as: $Z = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$


```python
circuit = QuantumCircuit(3)
circuit.x(0)
circuit.barrier()

circuit.y(1)
circuit.barrier()

circuit.z(2)

circuit.measure_all()
circuit.draw(output='mpl')
```

```python
state_vector(circuit)
```

```python
plot_bloch(circuit)
```

```python
aersim = AerSimulator()
# Perform an ideal simulation
result_ideal = aersim.run(circuit).result()
counts_ideal = result_ideal.get_counts(0)
plot_histogram(counts_ideal)
```

The **Hadamard gate** is named after the French mathematician Jacques Hadamard.
The Hadamard gate operates on a single qubit and transforms the basis states |0⟩ and |1⟩ into superposition states. In matrix form, the Hadamard gate is represented as:
$H = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}$
When applied to a qubit, the Hadamard gate transforms the states as follows:

*   $H|0\rangle =  \frac{1}{\sqrt{2}} (|0\rangle + |1\rangle)$

*   $H|1\rangle =  \frac{1}{\sqrt{2}} (|0\rangle - |1\rangle)$

The Hadamard gate creates an *equal superposition of the |0⟩ and |1⟩* states, resulting in an equal probability of measuring the qubit as |0⟩ or |1⟩. Additionally, the Hadamard gate is its own inverse, meaning that applying it twice returns the qubit to its original state. This self-inverting property is significant in various quantum algorithms.

```python
circuit = QuantumCircuit(3)

circuit.x(0)
circuit.h(0)
circuit.barrier()

circuit.h(1)
circuit.barrier()

circuit.h(2)
circuit.h(2)
circuit.barrier()

circuit.measure_all()
circuit.draw(output='mpl')
```

```python
state_vector(circuit)
```

```python
plot_bloch(circuit)
```

```python
aersim = AerSimulator()
# Perform an ideal simulation
result_ideal = aersim.run(circuit).result()
counts_ideal = result_ideal.get_counts(0)
plot_histogram(counts_ideal)
```

The S gate, also known as the phase gate, is a single-qubit quantum gate that introduces a phase shift on the |1⟩ state without affecting the |0⟩ state. In matrix form, the S gate is represented as: $S = \begin{bmatrix} 1 & 0 \\ 0 & i \end{bmatrix}$
When applied to a qubit, the S gate transforms the states as follows:

*   $S|0⟩ = |0⟩$

*   $S|1⟩ = i|1⟩$




```python
circuit = QuantumCircuit(2)

circuit.rx(pi/3,0)
circuit.s(0)
circuit.barrier()

circuit.rx(pi/3,1)
circuit.s(1)
circuit.s(1)
circuit.barrier()

circuit.measure_all()
circuit.draw(output='mpl')
```

```python
state_vector(circuit)
```

```python
plot_bloch(circuit)
```

```python
aersim = AerSimulator()
# Perform an ideal simulation
result_ideal = aersim.run(circuit).result()
counts_ideal = result_ideal.get_counts(0)
plot_histogram(counts_ideal)
```
